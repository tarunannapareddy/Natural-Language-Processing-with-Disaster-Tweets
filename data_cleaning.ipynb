```python
import pandas as pd
import matplotlib.pyplot as plt
import re



train_file = pd.read_csv('train.csv')


def clean_text(text):
    # Make the text all lower case
    text= text.lower()
    
    # Remove special characters and punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove extra white spaces (more than one space)
    text = re.sub(r'\s+', ' ', text)
    
    # Strip leading and trailing white spaces
    text = text.strip()
    
    return text

def clean_location(text):
    # Make the input a string
    text = str(text)
    
    # Make the text all lower case
    text = text.lower()
    
    # Remove special characters except for commas
    text = re.sub(r'[^\w\s,]', '', text)
    
    # Remove extra white spaces (more than one space)
    text = re.sub(r'\s+', ' ', text)
    
    # Strip leading and trailing white spaces
    text = text.strip()
    
    return text


def cleaner(file):
    file["text"] = file["text"].apply(lambda text: clean_text(text))
    file["location"] = file["location"].apply(lambda loc: clean_location(loc))
    return file

train_file = cleaner(train_file)

train_file.head(56)
```
